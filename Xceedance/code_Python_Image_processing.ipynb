# Image proccessing with python

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Images
    - collection of pixels = pixels arranged in matrices
    - Images are usually processed in 8 bit size, which means a pixel value can range from 0 to 255
    - Grayscale images are 1Dimensional, one pixel is represented by 1 number b/w 0 to 255
    - Color images are 3Dimensional, one pixel is represented by 3 numbers (Red, Green, Blue) each of them ranging between 0 to 255

```python
x = np.array([[0,255,0,255],
              [0,255,0,255],
              [0,255,0,255],
              [0,255,0,255]])
```

```python
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,255,0,255],
              [255,0,255,0],
              [0,255,0,255],
              [255,0,255,0]])

plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,50,100,125],
              [220,200,180,150],
              [240,255,0,255],
              [255,0,255,0]])

plt.imshow(x,cmap='gray')
plt.show()
```

#### Color images 
    - each pixel is composed of 3 values - RGB
    - [0,0,255] = Blue
    - [255,0,0] = Red
    - [0,255,0] = Green
    - [0,0,0] = Black
    - [255,255,0] = yellow

```python
y = np.array([ [ [255,0,0], [255,255,0]],
               [ [0,255,0], [255,120,50]],
               [ [0,0,255], [25,200,120]]])
plt.imshow(y)
plt.show()
```

# Morphological Analysis - erosion, dilation of images

```python
# erosion - loss of pixel values
# dilation - gain of pixel values
```

```python
img = plt.imread(r"C:\Users\aspdi\Downloads\image_data\image_data\np2.jpg")
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
kernel = np.ones((7,7))
# erosion
img2 = cv2.erode(img,kernel)
plt.imshow(img2)
plt.show()
```

```python
# dilation
img3 = cv2.dilate(img2,kernel)
plt.imshow(img3)
plt.show()
```

```python
img = plt.imread(r"C:\Users\aspdi\Downloads\image_data\image_data\np3.jpg")
plt.imshow(img)
plt.show()
```

## Working with haarcascade

```python
fd = cv2.CascadeClassifier(r"C:\Users\aspdi\Downloads\image_data\image_data\haarcascade_frontalface_alt.xml")
img = plt.imread(r"C:\Users\aspdi\Downloads\image_data\image_data\happy3.jpg")
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
corners = fd.detectMultiScale(img)
corners
```

```python
for (x,y,w,h) in corners:
    cv2.rectangle(img,[x,y],[x+w,y+h],[0,255,0],2)
```

```python
plt.imshow(img)
plt.show()
```

## Working with videos

```python
#create a video object
vid = cv2.VideoCapture(r"C:\Users\aspdi\Downloads\video3.mp4")

while True:
    flag,img = vid.read()
    if flag:
        cv2.imshow("Video",img)
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
vid.release()
cv2.destroyAllWindows()
```

```python
#create a video object
vid = cv2.VideoCapture(r"C:\Users\aspdi\Downloads\video3.mp4")

while True:
    flag,img = vid.read()
    if flag:
        
        corners = fd.detectMultiScale(img)
        for (x,y,w,h) in corners:
            cv2.rectangle(img,[x,y],[x+w,y+h],[0,255,0],2)
        cv2.imshow("Video",img)
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
vid.release()
cv2.destroyAllWindows()
```
