# Data Wrangling

```python
import pandas as pd
```

```python
# loading regiment.csv
df1 = pd.read_csv(r"C:\Users\aspdi\Downloads\datasets-1\datasets-1\regiment.csv",index_col='index')
df1.shape
df1
```

```python
# loading regiment2.csv
df2 = pd.read_csv(r"C:\Users\aspdi\Downloads\datasets-1\datasets-1\regiment2.csv")
df2.shape
df2
```

```python
df2.info()
```

```python
df = pd.merge(left=df1,right=df2,on='name',how='inner')

# df = pd.merge(left=df1,right=df2,left_on='name',right_on="name",how='inner')
```

```python
df
```

```python
# create a new column "improvement" - representing improvement made by soldiers during the training
df['improvement'] = df['postTestScore'] - df['preTestScore']
```

```python
df
```

```python
# convert all the names to capital case
df.name = df.name.apply(lambda x:x.upper())
df
```

```python
from datetime import datetime as dt
dt.today()
```

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth']) # standardizing the datetime col
```

```python
df.info()
```

```python
# create a new column age - calculated value from date of birth
df['Age'] = df.date_of_birth.apply(lambda x:round((dt.today()-x).days/365))
df
```
