# Data Visualization using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.5)
x
```

```python
y = x**2
z = x**3
```

```python
# line plot
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x,z)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y,c='r',label=' x vs y')
plt.plot(x,z,c='g',label=' x vs z')
plt.title(" graph of x v/s y")
plt.xlabel("values of x")
plt.ylabel("values of y")
plt.legend()
plt.show()
```

```python
#Scatter plot
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y,c='r')
plt.plot(x,y,c='g')
plt.show()
```

```python
# bar plot
names = ['BOM','DEL','HYD','BLR','PNQ','CCU','MAA']
vals = [45,56,34,56,49,67,46]

plt.figure(figsize=(12,5))
plt.bar(names,vals)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.bar(names,vals,color=['red','lightgreen','pink','yellow','orange','lightblue','green'])
plt.show()
```

```python
# pie chart
plt.figure(figsize=(10,10))
plt.pie(vals,labels=names)
plt.show()
```

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv(r"C:\Users\aspdi\Downloads\datasets-1\datasets-1\Bank_churn_modelling.csv")
df.shape
```

```python
df.head()
```

```python
fig = px.histogram(df,x='Geography')
fig.show()
```

```python
# scatter plot
fig = px.scatter(df,x='Age',y="Balance",color='Exited')
fig.show()
```

```python
# distplot
import plotly.figure_factory as ff
fig = ff.create_distplot([df.Age[df.Exited==0],df.Age[df.Exited==1]],['Not Exited','Exited'])
fig.show()
```

```python
df.columns
```

```python
df2 = df[['Geography','Age', 'Balance','IsActiveMember','Exited']]

fig = px.parallel_coordinates(df2.sample(3000),color='Exited')
fig.show()
```
